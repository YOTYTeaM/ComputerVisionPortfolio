```python
from deepface import DeepFace
import matplotlib.pyplot as plt
import cv2, json


img1_path = "Data/deepface/img1.jpg"
img2_path = "Data/deepface/img2.jpg"

result = DeepFace.verify(img1_path = img1_path, img2_path = img2_path)
print(type(result), json.dumps(result, indent=2))

plt.title(f"Is Verified: {result['verified']}")
plt.axis('off')

plt.subplot(1,2,1)
img1 = cv2.imread(img1_path)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img1)

plt.subplot(1,2,2)
img2 = cv2.imread(img2_path)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(img2)

plt.show()
```

```python
objs = DeepFace.analyze(
  img_path = img1_path, actions = ['age', 'gender', 'race', 'emotion']
)
print(type(objs), json.dumps(objs, indent=2))
```
